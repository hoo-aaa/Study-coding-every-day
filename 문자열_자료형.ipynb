```python
#포맷 이름으로 하기 
"I ate {number} apples, so I was sick for {day} days.".format(number=10,day=3)
```

```python
#왼쪽 정렬 
"{0:<10}".format("안녕")
#:<10 표현식을 사용하면 문자열을 왼쪽으로 정렬하고 종 자릿수를 10으로 맞춰준다. 
```

```python
"{0:>10}".format("안녕")

#오른쪽 정렬 
```

```python
"{0:^10}".format("안녕")
#가운데 정렬 
```

```python
"{0:=^10}".format("안녕") #설명: 가운데 정렬을 하되 빈 공간을 =로 채워줘
```

```python
y = 3.42145234
"{0:0.4f}".format(y)
```

```python
"{0:10.4f}".format(y) #길이를 10자리로 만들고 소숫점은 4개 뽑아줘 10.4f 
```
